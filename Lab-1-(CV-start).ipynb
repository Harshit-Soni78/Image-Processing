```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_image = cv2.imread("demo.jpg",cv2.IMREAD_GRAYSCALE)
```

```python
input_image = cv2.resize(input_image, (1280, 800))
```

```python
# Display the image in a window
cv2.imshow('Image', input_image)

# Wait for a key press and close the window
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(input_image, cmap="gray", interpolation="bicubic")
plt.show()
```

```python
plt.imshow(input_image, cmap="gray", interpolation="bicubic")
plt.plot([500,100],[800,100],'c', linewidth=5)
plt.show()
```

```python
cv2.imwrite('gray_demo.png', input_image)
```
